```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
patient1 = [76,126,38.0]
patient2 =[74,120,38.0]
patient3 = [72,118,37.5]
patient4 = [78,136,37.0]
```

```python
patients = [patient1, patient2, patient3, patient4]
data = np.array(patients)
print(patients)
data
```

**1.Which patient is farthest from the rest?**

```python
import numpy as np

num_points = data.shape[0]

max_distance = 0
farthest_patient_index = None

farthest_patient_pair = max(
    ((i, j, np.linalg.norm(data[i] - data[j])) for i in range(num_points) for j in range(i + 1, num_points)),
    key=lambda x: x[2]
)

farthest_patient_index = farthest_patient_pair[1]

print(f"Patient Has the  farthest distance {farthest_patient_index + 1}")
print(f"distance is: {farthest_patient_pair[2]}")
```

**2. Which two patients are nearest?**


```python

num_patients = data.shape[0]
min_distance = float('inf')
patient1_index, patient2_index = -1, -1

# Calculate distances and find nearest patients using for loops
for i in range(num_patients):
    for j in range(i + 1, num_patients):
        distance = np.linalg.norm(data[i] - data[j])
        if distance < min_distance:
            min_distance = distance
            patient1_index, patient2_index = i, j

print(f"Patient {patient1_index + 1} and Patient {patient2_index + 1} are nearest.")
print("Distance:", min_distance)
```

**3. Create a new dummy patient, and then find which is the closest patient?**

```python
import numpy as np


#dummy pateint
num_patients = data.shape[0]
new_dummy_patient = np.array([76, 136, 36.8])

distances = np.zeros((num_patients, num_patients))
for i in range(num_patients):
    for j in range(num_patients):
        distances[i, j] = np.linalg.norm(data[i] - data[j])

new_patient_distances = np.zeros(num_patients)

# Calculate distances to the new dummy
for i in range(num_patients):
    new_patient_distances[i] = np.linalg.norm(data[i] - new_dummy_patient)

closest_patient_index = np.argmin(new_patient_distances)

print(f"The closest patient to the new dummy patient is Patient {closest_patient_index + 1}")
print("Distance:", new_patient_distances[closest_patient_index])
```
